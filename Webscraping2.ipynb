```python
import pandas as pd
df = pd.DataFrame(columns=['link','title','company_link','company','job_type'])

def parse_content(adverts):
  link = []
  title = []
  company = []
  company_link = []
  job_type = []
  for advert in adverts:
    link.append('https://www.jobinrwanda.com' + advert.find('a')['href'])
    title.append(advert.find('span').getText())
    company.append(advert.find('p').find('a').getText())
    company_link.append(advert.find('p').find('a')['href'])
    job_type.append(advert.find('p').find('span').getText())
  return link, title, company, company_link, job_type
```

```python
link = 'https://www.jobinrwanda.com'
content = requests.get(link).content
raw_content = BeautifulSoup(content, 'html.parser').find_all('div', class_='card-body p-2')
df['link'], df['title'], df['company'], df['company_link'], df['job_type'] = parse_content(raw_content)
df
```

```python
def get_description(job_link):
  job_description_raw_content = requests.get(job_link).content # request raw content
  job_description_bs = BeautifulSoup(job_description_raw_content, 'html.parser')
  employer_description = job_description_bs.find_all('div', class_='employer-description')[0].getText()

  describe = job_description_bs.find_all('div', class_='clearfix text-formatted field field--name-field-job-full-description field--type-text-long field--label-hidden field__item')
  return employer_description
```

```python
# Adding description
descr = []
for index in range(len(df)):
  try:
    descr.append(get_description(df['link'][index]))
  except IndexError:
    descr.append('NA')
df['description'] = descr
df
```
